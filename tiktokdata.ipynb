```python
#Now Let's go on SQL
!pip install pandasql
import pandasql as ps
```

```python
# Top 10 most viewed video
query = """ SELECT video_id, video_view_count
FROM df_final
ORDER BY video_view_count DESC
LIMIT 10;
"""
result = ps.sqldf(query, locals())

print(result)
```

```python
# Global average engagemenrt rate
query = """ SELECT ROUND(AVG((video_like_count + video_comment_count + video_share_count + video_download_count) / video_view_count), 2) 
                AS avvg_engagement_rate
            FROM df_final
            WHERE video_view_count > 0;
"""
result = ps.sqldf(query, locals())

print(result)
```

```python
# Average length of video
query = """
SELECT ROUND(AVG(video_duration_sec),2) AS avg_video_duration
FROM df_final;
"""
result = ps.sqldf(query, locals())

print(result)
```

```python
# Number of video with more than million views
query = """
SELECT COUNT(*) AS video_over_million
FROM df_final
WHERE video_view_count > 1000000;
"""
result = ps.sqldf(query, locals())

print(result)
```

```python
# Top 5 video with best engagement rate
query = """
SELECT video_id, ROUND((video_like_count + video_comment_count + video_share_count + video_download_count)/ video_view_count, 2) AS engagement_rate
FROM df_final
WHERE video_view_count > 0
ORDER BY engagement_rate DESC
LIMIT 5;
"""
result = ps.sqldf(query, locals())

print(result)
```

```python
# Average length of video per slice of views
query = """
SELECT
    CASE 
        WHEN video_view_count < 10000 THEN '0-10k'
        WHEN video_view_count BETWEEN 10000 AND 100000 THEN '10k-100k'
        WHEN video_view_count BETWEEN 1000000 AND 1000000000 THEN '100k-1M'
        ELSE '+1M'
    END AS view_range,
    ROUND(AVG(video_duration_sec),2) AS avg_duration
FROM df_final
GROUP BY view_range
ORDER BY 
    CASE 
        WHEN view_range = '0-10k' THEN 1
        WHEN view_range = '10-100k' THEN 2
        WHEN view_range = '100k-1M' THEN 3
        ELSE 4
    END;  

        
"""
result = ps.sqldf(query, locals())

print(result)
```

```python
# Average engagement rate per slice of views
# same as before but by adding condition of the view count and adding engagement rate
query = """
SELECT
   CASE 
       WHEN video_view_count < 10000 THEN '0-10k'
       WHEN video_view_count BETWEEN 10000 AND 100000 THEN '10k-100k'
       WHEN video_view_count BETWEEN 100001 AND 999999 THEN '100k-1M'
       ELSE '+1M'
   END AS view_range,
   ROUND(AVG((video_like_count + video_comment_count + video_share_count + video_download_count) * 1.0 / video_view_count), 2) AS avg_engagement_rent
FROM df_final
WHERE video_view_count > 0
GROUP BY view_range
ORDER BY 
   CASE 
       WHEN view_range = '0-10k' THEN 1
       WHEN view_range = '10k-100k' THEN 2
       WHEN view_range = '100k-1M' THEN 3
       ELSE 4
   END;  
"""
result = ps.sqldf(query, locals())

print(result)
```
